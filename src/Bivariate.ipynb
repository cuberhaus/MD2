# Imports

```python
import pandas
import matplotlib.pyplot as plt
from pandas import plotting
import numpy as np
import seaborn as sns
```

# Read the Data

```python
data = pandas.read_csv(r'../data/tommaso_data/allUsers.lcl.csv', na_values='?')
#display(data[0:10])
```

```python
print("Correlations between numeric columns:")
display(data.corr(method='pearson'))
#data.dtypes
```

```python
print("Correlation between numeric columns:")
display(data.corr(method='pearson'))

correlation_matrix = data.corr(method='pearson')

# Sort the correlations in descending order.
sorted_correlations = correlation_matrix.abs().unstack().sort_values(ascending=False)

# Filter out correlations that are not (autocorrelations).
sorted_correlations = sorted_correlations[sorted_correlations != 1]

# Display the columns with the highest correlation.
print(sorted_correlations)

# Filter high correlations
high_correlations = correlation_matrix.abs() > 0.60

# Display high correlations
print("High Correlations:")
for column in high_correlations.columns:
    for index in high_correlations.index:
        if high_correlations.loc[index, column] and index != column:
            print(f"Correlation between {index} and {column} is high.")

# Heatmap of high_correlations Matrix
# Calculate correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-.8, vmax=.8, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

# Change the ticks color to white
plt.rcParams['ytick.color'] = 'white'
plt.rcParams['xtick.color'] = 'white'

plt.title('Correlation Matrix Heatmap')
plt.show()
```

# Scatter Plot

```python
# Choose columns for x, y, z coordinates.
x_columns = [f'X{i}' for i in range(12)]
y_columns = [f'Y{i}' for i in range(12)]
z_columns = [f'Z{i}' for i in range(12)]

# Loop through each pair of X and Y columns and plot a scatter plot
for x, y in zip(x_columns, y_columns):
    plt.figure(figsize=(10, 8))
    plt.scatter(data[x], data[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'Scatter plot between {x} and {y}')
    plt.show()
```

# Distribution of Classes

```python
data['Class'] = data['Class'].astype(str)

plt.rcdefaults()
plt.figure(figsize=(10, 8))
sns.countplot(x='Class', data=data, palette='viridis')

# Show values on top of bars
for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width()/2., height + 0.1,
                   '{}'.format(int(height)), ha='center', va='bottom')

plt.title('Distribution of Classes', size=20)
plt.xlabel('Class', size=16)
plt.ylabel('Count', size=16)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()

# Print out the counts for each class
class_counts = data['Class'].value_counts()
print(f'Counts for each class:\n{class_counts}')
```

# Marker Positions with Respect to Classes

```python
marker_columns = ['X0', 'Y0', 'Z0']

for col in marker_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Class', y=col, data=data)
    plt.title(f'Boxplot of {col} grouped by Class')
    plt.show()
```

# Correlation Heatmap of Markers

```python
# Calculate correlation matrix
corr = data[marker_columns].corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Matrix Heatmap of Marker Positions')
plt.show()
```

# Pairwise Scatter Plots for Each Class

```python
for class_type in data['Class'].unique():
    # Filter data for the current class
    class_data = data[data['Class'] == class_type]
    
    # Create pairwise scatter plot for the first three markers
    sns.pairplot(class_data[marker_columns])
    plt.suptitle(f'Pairwise Scatter Plots for Class {class_type}', y=1.02)
    plt.show()
```
